```python
import tensorflow as tf
from tensorflow.keras import datasets,models,layers
import matplotlib.pyplot as plt
```

```python
(train_images,train_labels),(test_images,test_labels)=tf.keras.datasets.mnist.load_data()
train_imges=train_images.reshape(-1,28,28)
test_imges=test_images.reshape(-1,28,28)
```

```python
train_images,test_images=train_images/255.0,test_images/255.0
```

```python
model =models.Sequential()
model.add(layers.SimpleRNN(32, activation="relu", input_shape=(28, 28), return_sequences=True))
model.add(layers.SimpleRNN(16, activation="relu"))
model.add(layers.Dense(10))
```

```python
model.compile(optimizer="adam",loss="mse",metrics=['accuracy'])
```

```python
history=model.fit(
    train_images,train_labels,
    epochs=5,
    validation_data=(test_images,test_labels)
    )
```
